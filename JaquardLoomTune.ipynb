```python
%pylab inline
from __future__ import print_function
from __future__ import division
from pylab import *
img=imread('/Users/alexisstorycrawshaw/Documents/Documents/201A/Jacquard_Fabric.jpg')
imshow(img)
```

```python
imshow(img[:,:,0],cmap=cm.gray)
```

```python
sig= img[100, 30:460, 1]
plot(sig);
sig= img[100, :, 1]
plot(sig);
```

```python
sigout = list((sig.astype(int16) - 127) * 200)*300
plot(sigout)
```

```python
from scipy.io import wavfile
wavfile.write('outsig.wav', 44100, array(sigout, dtype=int16))
```

```python
newSigOut = array(((img[:,:,0].astype(int16)-170)*200).flat)
plot(newSigOut)
```

```python
from scipy.io import wavfile
wavfile.write('outsig3.wav', 44100, array(newSigOut, dtype=int16))
```

```python
longSig = array([newSigOut,newSigOut,newSigOut]).flat
wavfile.write('outsigLong44100.wav', 44100, array(longSig, dtype=int16))
wavfile.write('outsigLong8000.wav', 8000, array(longSig, dtype=int16))
```
