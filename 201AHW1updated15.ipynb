```python
%pylab inline
from __future__ import print_function
from __future__ import division

import csv

yearArray = []
maleArray = []
femaleArray = []

with open('/Users/alexisstorycrawshaw/LifeExpect.csv', 'U') as f:
    reader = csv.reader(f)
    for row in reader:
        yearArray.append(row[0])
        maleArray.append(row[1])
        femaleArray.append(row[2])
    
#print yearArray
#print maleArray
#print femaleArray

plot(yearArray, maleArray, ':')
#axis(yearArray)
plot(yearArray, femaleArray, ':')
xlabel('Year', fontsize=8)
ylabel('Life Expectancy (years)', fontsize=8)
title('Life Expectancy in British Columbia, Canada by Year (for a Newborn)', fontsize=10)

names = ['Male', 'Female']
legend(names)
savefig('201AHW1_graph.pdf')
```
